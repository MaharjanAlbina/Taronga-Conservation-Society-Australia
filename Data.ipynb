```python
import pandas as pd

# Replace with your actual file path or name
file_path = '/Users/albina/Documents/Internship/Compiled Data-2.xlsx'

# Read the Excel file
df = pd.read_excel(file_path, engine='openpyxl')

# Print the contents of the file
print(df)

# Optionally, display column names
print("Columns:", df.columns.tolist())
```

```python
# Safely convert Excel date format (serial numbers) to datetime
df['Date'] = pd.to_numeric(df['Date'], errors='coerce')  # ensure numeric
df['Date'] = pd.to_datetime(df['Date'], unit='D', origin='1899-12-30', errors='coerce')

# Convert Start and End Time if still in Excel float format
df['Start Time'] = pd.to_numeric(df['Start Time'], errors='coerce')
df['End Time'] = pd.to_numeric(df['End Time'], errors='coerce')
df['Start Time'] = pd.to_datetime(df['Start Time'], unit='D', origin='1899-12-30', errors='coerce').dt.time
df['End Time'] = pd.to_datetime(df['End Time'], unit='D', origin='1899-12-30', errors='coerce').dt.time
```

```python
df
```

```python
import numpy as np

# Clean columns and prepare for update
df.columns = df.columns.str.strip()
df['Species'] = df['Species'].astype(str).str.strip()
df['Activity Description'] = df['Activity Description'].astype(str).str.lower()

# Define keyword-to-species mapping
keywords = {
    'meerkat': 'Meerkat',
    'lion': 'Lion',
    'sun bear': 'Sun Bear',
    'bear': 'Sun Bear',
    'tiger': 'Tiger',
    'otter': 'Otter',
    'binturong': 'Binturong',
    'fox': 'Fennec Fox',
    'fennec': 'Fennec Fox',
    'red panda': 'Red Panda',
    'chicken': 'Chicken'
}

# Function to infer species
def infer_species(desc):
    for key, species in keywords.items():
        if key in desc:
            return species
    return np.nan

# Detect missing species
missing_species_mask = df['Species'].isin(['', '–', 'nan', 'NaN', 'None'])

# Update 'Species' column in the same dataframe
df.loc[missing_species_mask, 'Species'] = df.loc[missing_species_mask, 'Activity Description'].apply(infer_species)

# Done: df is now updated
print("Species column updated inside the existing df.")
```
